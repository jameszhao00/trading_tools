# Study FB's prophet on synthetic data

## Takeaways
- Doesn't handle mean reversion well. (but we don't really need mean reversion)

## Helper functions

```python

import pandas as pd
from fbprophet import Prophet
import numpy as np
import matplotlib.pylab as plt
```

```python
def fuzz(d):
    std = np.std(d)
    return d + np.random.uniform(-std/5.0, std/5.0, size=d.size)
fuzz(np.arange(10))
```

```python
def study_prophet(x, y, show_components=False):
    data = pd.DataFrame({
        "y": y, 
        "ds": x
    })
    m = Prophet(yearly_seasonality=False, weekly_seasonality=False, daily_seasonality=False,
               interval_width=0.95)
    m.fit(data)
    future = m.make_future_dataframe(periods=365)
    forecast = m.predict(future)
    m.plot(forecast)
    if show_components:
        m.plot_components(forecast)
    return forecast
```

```python
date_range = pd.date_range(start='1/1/2018', periods=1000, freq='1D')
```

## Studies

```python
# Sine wave. No prediction because we turned off all seasonality.
p = study_prophet(date_range, fuzz(np.sin(pd.Series(date_range.values).dt.dayofyear/50.0)))
```

```python
p.set_index("ds")
```

```python
# Strong Trend with random noise.
study_prophet(date_range, fuzz(np.arange(date_range.size)))
```

```python
# Random walk with a very slight uptrend.
study_prophet(date_range, (np.random.uniform(-.01, .0103, size=date_range.size)).cumsum(), 
              show_components=True)
```

```python
# Generate a momentum/mean reverting series with moving average.
import statsmodels.tsa.arima_process
def gen_series(arparams, maparams, size):
    ar = np.r_[1, -arparams] # add zero-lag and negate
    ma = np.r_[1, maparams] # add zero-lag
    d = statsmodels.tsa.arima_process.arma_generate_sample(ar, ma, size)
    return d
```

```python
# Momentum.
study_prophet(date_range, gen_series(np.array([1.0001, 0]), np.array([0, 0]), size=date_range.size), 
              show_components=True)
```

```python
# Mean reverting.
study_prophet(date_range, gen_series(np.array([.99, 0]), np.array([0, 0]), size=date_range.size), 
              show_components=True)
```

```python
study_prophet(date_range, fuzz(np.arange(date_range.size)), show_components=True);
```
